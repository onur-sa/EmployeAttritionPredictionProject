# Problem Definition and Objectives
Today, one of the most important problems of companies is the inability to retain a qualified workforce. A huge company with thousands of employees has a hard time finding out if each employee is worn out from the job. For this, companies need to be able to understand the wishes of their employees and the difficult mental and physical situations they are in. This is a huge workload and financial burden.

This project was done to create an AI model that predicts employee attrition with various information from employees. Using this information, companies can decide what improvements they can make to provide a better working environment for their employees and to look after their employees' happiness.
Thus, they can identify the employees who are likely to quit their job and solve the problems of these employees or offer some promotions to keep them from leaving the job.

## One Hot Encoding
- With One Hot Encoding, we converted each category of features containing categorical information into a new feature, and we did this conversion process with binary numbers. In this way, the number of features increased to 54.
- OHE generally recommended when data is not sorted.
*
- One Hot Encoding ile kategorik bilgiler içeren featureların her bir kategorisi yeni bir feature'a dönüştürdük bu dönüştürme işlemini binary sayılarla yaptık. Bu şekilde feature sayısı 54 e çıktı.
- OHE Genelde veriler sıralı değilken önerilir.

```python
columns_to_encode = ["BusinessTravel","Department","EducationField","Gender","JobRole","MaritalStatus","OverTime"]
```

```python
## Sorun çıkardığı için BusinessTravel 'ı ayrıca encode ettim.
## I encoded BusinessTravel seperately becouse there was some problem.
train_cleaned["BusinessTravel"] = train_cleaned["BusinessTravel"].map({"Travel_Rarely":1, "Travel_Frequently":2, "Non-Travel":0})
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop="if_binary")
encoder_df = pd.DataFrame(encoder.fit_transform(train_cleaned[columns_to_encode]).toarray())
column_names = encoder.get_feature_names(columns_to_encode)
column_names
```

```python
encoder_df.columns = column_names
```

```python
train_cleaned.drop(columns=columns_to_encode, inplace=True, axis=1)
train_cleaned = train_cleaned.join(encoder_df)
```

## Error Metrics


```python
from sklearn import metrics

def eval_clf(y_true,y_pred):
    acc = metrics.accuracy_score(y_true,y_pred)
    p = metrics.precision_score(y_true,y_pred)
    r = metrics.recall_score(y_true,y_pred)
    f1= metrics.f1_score(y_true, y_pred)
    
    result = {"acc":acc,
             "p": p,
             "r": r,
             "f1": f1}
    return result

def eval_auc(y_true, y_prob):
    fpr,tpr, threshold = metrics.roc_curve(y_true, y_prob)
    roc_auc = metrics.auc(fpr,tpr)
    return roc_auc

def eval_threshold(y_true, y_prob, threshold):
    y_pred = np.where(np.array(y_prob) > threshold, 1,0)
    
    return eval_clf(y_true,y_pred)
```

- I created the functions evaluation classification, evaluation auc, evaluation threshold. After making the predictions, I will be able to call these functions and easily evaluate the model performance.
- evaluation classification, evaluation auc, evaluation threshold fonksiyonlarını oluşturdum. Tahminleri yaptıktan sonra bu fonksyonları çağırıp kolayca model performansını değerlendirebileceğim.

## Hyper Parameter Optimization

* The LogisticRegression() function has many parameters and some of these parameters have numeric values. In this case, we need to do a fair amount of experimentation to run the function with the best parameters. Hyper Parameter Optimization does these trials for us. We assign various values to each parameter that we will use in the Logistic Regression model. HPO tries all combinations of these values and finds the most suitable combination for us. I found the parameters of my Logistic regression model using Hyper Parameter Optimization.
*
* LogisticRegression() fonksiyonun birçok parametresi var ve bu parametrelerden bazıları sayısal değere sahip. Bu durumda fonksiyonu en iyi parametrelerle çalıştırmak için oldukça fazla miktarda deneme yapmamız gerekiyor. Hyper Parameter Optimization bu denemeleri bizim için yapıyor. Logistic Regresyon modelinde kullanacağımız herbir parametreye çeşitli değerler atıyoruz. HPO bizim için bu değerlerin bütün kombinasyonlarını deneyip en uygun olan kombinasyonu buluyor. Ben Lojistik regresyon modelimin parametrelerini Hyper Parameter Optimization kullanarak buldum.
